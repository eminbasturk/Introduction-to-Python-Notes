# MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

**1)** <code>plt.plot(x, y)</code> **Line Plot** oluşturmaya yarayan fonksiyondur. Zaman ve ardışıl seriler için kullanılır. Genelde **x**'e zaman, **y**'ye de değer koyulur.

```python
plt.plot([1998, 2015, 2021], ["E", "S", "Y"]);
```

**2)** <code>plt.show()</code> Çizilen grafiği gösteren fonksiyon.

```python
plt.show()
```

**3)** <code>plt.scatter(x, y, c, s, alpha)</code> **Scatter Plot** oluşturan fonksiyondur. İki değer arasındaki korelasyonu karşılaştırmak için kullanılır. **x** ve **y** değerler.
* <code>c</code>: grafiği renklendiren argüman
* <code>s</code>: grafiği boyutlandıran argüman
* <code>alpha</code>: grafiği şeffaflaştıran argüman

```python
plt.scatter([1, 2, 3], [3, 2, 1]);
```

**4)** <code>plt.xscale("log")</code> **x** ekseninin ölçeğini logaritmik olarak belirleyen fonksiyon.

```python
plt.scatter([1, 2, 3], [3, 2, 1])
plt.xscale("log")
plt.show()
```

**5)** <code>plt.hist(x)</code> **Histogram** oluşturan fonksiyon. **x** arrayinin değer dağılımını gösterir.
* <code>bins</code> grafiğin kaç sütundan oluşacağını belirleyen argüman

```python
plt.hist([1, 2, 3, 4, 5, 3, 3, 2, 4, 4, 4, 4, 1, 1], bins = 20)
plt.show()
```

**6)** <code>plt.clf()</code> Önceki grafiği silen fonksiyon.

```python
plt.clf()
```

**7)** <code>plt.xlabel()</code>, <code>plt.ylabel()</code> ve <code>plt.title()</code> **x** ve **y** eksenlerinin etiketlerini belirlemeye ve başlık belirlemeye yarayan fonksiyonlar.

```python
plt.scatter([1, 2, 3], [3, 2, 1])
plt.xlabel("Sayılar")
plt.ylabel("Rakamlar")
plt.title("Sayılar ve Rakamlar")
plt.show()
```

**8)** <code>plt.xticks(ticks, labels)</code> ve <code>plt.yticks(ticks, labels)</code> **x** ve **y** eksenlerini etiketleyen fonksiyonlar.

**9)** <code>plt.text(x, y, s)</code> Değerlere göre grafik içine string yazan fonksiyon.
* <code>x</code> x ekseninin değer argümanı
* <code>y</code> y ekseninin değer argümanı
* <code>s</code> stringin yazıldığı argüman

**10)** <code>plt.grid(b)</code> Grafiği çizgi çizgi gösteren fonksiyon
* <code>b</code> argümanı **True** olmalıdır.

# DICTIONARIES & PANDAS

### Dictionaries

```python
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }
```

**11)** <code>.keys()</code> Sözlüğün keylerini veren metot.

```python
europe.keys()
```

**12)** <code>.values()</code> Sözlüğün değerlerini veren metot.

```python
europe.values()
```

**13)** <code>.items()</code> Sözlüğün key-value ikililerini veren metot.

```python
europe.items()
```

### Pandas

```python
import pandas as pd
```

**13)** <code>pd.DataFrame(data)</code> İçine verilen **data** ile **DataFrame** oluşturan fonksiyon.
* <code>index</code> DataFrame'in indexini manuel olarak ayarlamaya yarayan argüman.

```python
countries = pd.DataFrame(europe, index = [0])
countries
```

**14)** <code>.index</code> DataFrame'in indexlerini veren özellik.

```python
countries.index
```

**15)** <code>pd.read_csv("data.csv")</code> CSV dosyalarıın DataFrame olarak okumaya yarayan fonksiyon.
* <code>index_col</code> DataFrame'in hangi sütununun index olacağını belirleyen argüman.

**16)** <code>.loc[]</code> DataFrame içinden etiketlere göre seçim yapmaya yarayan teknik.

```python
countries.loc[:,"spain":"france"]
```

**17)** <code>.iloc[]</code> DataFrame içinden indexlere göre seçim yapmaya yarayan teknik.

```python
countries.iloc[:, 0:2]
```

# LOOPS

**21)** <code>range()</code> For döngüsünün ne kadar döneceğini belirleyen fonksiyon.

```python
for i in range(1, 10):
    print(i)
```

**22)** <code>enumerate()</code> Döngüler içinde nesneleri indexleri ile döndüren fonksiyon.

```python
for i, d in enumerate(["A","B","C"]):
    print(i, d)
```

**23)** <code>np.nditer()</code> 2D Numpy arraylerini tek tek yazdırmak için kullanılan döngü fonksiyonu.

**24)** <code>.iterrows()</code> DataFrameler arasında gezinmeyi sağlayan metot. İlk değişken satırı, ikinci değişken sütunu verir.

```python
for i, d in countries.iterrows():
    print(i)
    print(d)
```
